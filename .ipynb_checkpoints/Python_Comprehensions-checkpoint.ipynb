### ------------- List Comprehension -----------
List: List is collection of data surrounded by square brackets and each elements are seperated by comma.
List Comprehension: Is also surrounded by square brackets but instead of list of data inside it contain expression like for loop followed by if clauses.

- [ expr for val in collection]
- [expr for val in collection if <test>]
- [expr for val in collection if <test1> and <test2>]
- [expr for val1 in collection1 and val2 in collection2]

```python
## List of square of first 100 postive number
squares = []
for i in range(1,101):
    squares.append(i**2)
    
print(squares)
```

```python
Squares2 = [i**2 for i in range(1,101)]
print(Squares2)
```

```python
## Complex examples
remainder5 = [x**2 % 5 for x in range (1,101)]
print(remainder5)
```

```python
remainders11 = [x**2 %11 for x in range(1,110)]
print(remainders11)
```

## Problem - Quadratic Reciprocity - proved by Gauss
P_remainders = [X**2 % P for x in range(0,p)]
- prime number(P) the number of remainders i.e. len(p_remainders) = P+1/2


```python
## List comprehension with if clause
### List of movies start with letter G
movies = ['Star Wars','Glass','The Upside','The LEGO Movie 2','Cold Pursuit',
         'Hotel Mumbai']

# if not using list comprehension
Gmovies = []
for title in movies:
    if title.startswith("G"):
        Gmovies.append(title)
        
print(Gmovies)
```

```python
## Using list comprehension
Gmovies = [title for title in movies if title.startswith("G")]
print(Gmovies)
```

```python
## Suppose list consist tuples with movie name and year released
MoviesYear = [('Star Wars',2019),('Glass',2020),('The Upside',2018),('The LEGO Movie 2',2020),('Cold Pursuit',2017),
         ('Hotel Mumbai',2020)]

TitleWithYears = [title for (title, year) in MoviesYear if year>2019]
print(TitleWithYears)
```

```python
## Some mathematical applications
Vector = [2,-3,1]
#Scalar multiplication
4*Vector # give 4 copy of list
```

```python
ScalarMultiplication  = [4*X for X in Vector]
print(ScalarMultiplication)
```

### Cartesian product --> Rene Descartes
- If A and B are sets Then cartesian products is set of pairs (a,b)



```python
A = [1,3,5,7]
B = [2,4,6,8]
CartesianProduct = [(a,b) for a in A for b in B]
print(CartesianProduct)
```

### ----------------------- Dictionary Comprehension ---------------------------
#### Dictionary:
- A dictionary is a collection which is unordered, changeable and indexed. In Python written with curly brackets, and they have keys and values.


```python
DictionaryExample = {"IDE": "JupyterNotebook", "Language": "Python", "Version": 3}
print(DictionaryExample)
```

```python
## Dictionary comprehension example
# Suppose we have to check the count of letters/alphabets repeated in defination of dictionary.

DictDefination = "A dictionary is a collection which is unordered, changeable and indexed. In Python written with curly brackets, and they have keys and values."

AlphabetDictionary = {key: DictDefination.count(key) for key in DictDefination}
print(AlphabetDictionary)
```

```python
#Lets plot
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

key = AlphabetDictionary.keys()
values =AlphabetDictionary.values()
plt.bar(key,values)
plt.xlabel('Alphabets')
plt.ylabel('Count')
```

```python
### Example cube of number i.e. key:
NumSquareDict = {1:1 , 2:8, 3:27}
## Dictionary comprehension
NumSquareDictComprehension = {f"The square of {key} is":key**3 for key in range(1,10)}
print(NumSquareDictComprehension)
```

### ------------------- Set Comprehension ---------------
- Set is collection of unique, unordered, mutable collection of elements. In python set elements written enclosed with curly brackets and elements seperated by comma.

```python
SquareSet = {num**2 for num in range (1,10)}
print(SquareSet)
```

```python
# Suppose list having duplicate elements
CarBrands = ['BMW','Chevrolet','Bentley','BMW']
print(CarBrands)
# Set comprehension only give only unique elements from list
CarBrandDict = {Brand for Brand in CarBrands}
print(CarBrandDict)
```

### --------------- Generator Expression Comprehension -----------

- Generator Expression:
